```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import svm  
from sklearn.multiclass import OneVsRestClassifier
from sklearn.multiclass import OneVsOneClassifier
```

```python
iris_data= pd.read_csv('iris.csv')
iris_data.head()
```

```python
# Plot the sepal features
sns.lmplot(' Sepal_Width', 'Sepal_Length', data=iris_data, hue='Class',
           palette='Set1', fit_reg=False);

sns.lmplot('Sepal_Length', ' Sepal_Width', data=iris_data, hue='Class',
           palette='Set1', fit_reg=False);
```

```python
# Plot the petal features
sns.lmplot(' Petal_Width', ' Petal_Length', data=iris_data, hue='Class',
           palette='Set1', fit_reg=False);

sns.lmplot(' Petal_Length', ' Petal_Width', data=iris_data, hue='Class',
           palette='Set1', fit_reg=False);
```

```python
X = iris_data.iloc[:,:4].values
Y = iris_data.iloc[:,4].values


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y)
print(X_train)
```

```python
print("Target labels : \n")
print(Y)
```

```python
print(f'Shape X_train: {X_train.shape}')
print(f'Shape y_train: {y_train.shape})')
print(f'Shape X_test: {X_test.shape}')
print(f'Shape y_test: {y_test.shape}')
```

```python
#OneVsRest Classifier
result_all=OneVsRestClassifier(svm.SVC(random_state=0)).fit(X_train, y_train)
result_all.predict(X_test)
```

```python
y_test
```

```python
#accuracy of one vs rest classifier
result_all.score(X_test, y_test)
```

```python
#One vs One classifier
result_one=OneVsOneClassifier(svm.SVC(random_state=0)).fit(X_train, y_train)
result_one.predict(X_test)
```

```python
#accuracy of one vs one classifier
result_one.score(X_test,y_test)
```
